#Print first five records from each collection

```python
import sys, re, pymongo, json, pprint
import requests

client = pymongo.MongoClient (host="da0.eecs.utk.edu")
# Get a reference to a particular database
db = client ['test']
names = db.collection_names()
output_file = open('db_sample_records.txt', 'w')
pp = pprint.PrettyPrinter(indent=2)

for name in names:
    print(name)
    output_file.write(name + '\n')
    print('**************************')
    output_file.write('**************************\n')
    
    coll = db[name]
    for record in coll.find({})[0:5]:
        pp.pprint(record)
        output_file.write(pp.pformat(record) + '\n')
```

#Get Schema Information


```python
import sys, re, pymongo, json, pprint
import requests

client = pymongo.MongoClient (host="da0.eecs.utk.edu")
# Get a reference to a particular database
db = client ['test']
names = db.collection_names()
output_file = open('db_collection_schema.json', 'w')

schema = {}

for name in names:
    
    schema[name] = {'count': 0, 'keys': [], 'mostRecentRecord': ''}
    uniqueKeys = {}
    
    coll = db[name]
    for record in coll.find({}).limit(5):
        for key in record:
            uniqueKeys[key] = 0
    for key in uniqueKeys:
        schema[name]['keys'].append(key)
    
    schema[name]['keys'].sort()
    schema[name]['count'] = coll.find().count()
    
schema = json.dumps(schema, sort_keys=True, indent=4)
print(schema)
output_file.write(schema + '\n')        
```
